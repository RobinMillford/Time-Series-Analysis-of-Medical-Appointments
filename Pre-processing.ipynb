```python
import pandas as pd
```

```python
df = pd.read_csv("Hospital_patients_datasets.csv")
df.head()
```

```python
df.duplicated().sum()
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df['ScheduledDay'] = pd.to_datetime(df['ScheduledDay']).dt.date
df['AppointmentDay'] = pd.to_datetime(df['AppointmentDay']).dt.date
```

```python
df.head()
```

```python
# Rename specific columns
df.rename(columns={'Hipertension': 'Hypertension',
                         'Handcap': 'Handicap',
                         'SMS_received': 'SMSReceived',
                         'No-show': 'NoShow'},
                inplace=True)
```

```python
# Drop specific columns
df.drop(columns=['PatientId', 'AppointmentID', 'Neighbourhood'], inplace=True)
```

```python
df.head()
```

```python
# Drop rows where 'Age' column is equal to 0
df = df[df['Age'] != 0]
    
# Define age group labels and age bins
age_labels = ['0-19', '20-39', '40-59', '60-79', '80+']
age_bins = [0, 19, 39, 59, 79, 150]  # Using 150 as an upper bound
    
# Categorize 'Age' values into bins using pd.cut()
df['Ag_group'] = pd.cut(df['Age'], bins=age_bins, labels=age_labels, right=False)
```

```python
df.head()
```

```python
df.drop(columns=['Age'], inplace=True)
```

```python
# Map 'NoShow' column values to binary format
df['NoShow'] = df['NoShow'].map({'Yes': 1, 'No': 0})
```

```python
df.head()
```

```python
df.to_csv('patients.csv', index=False)
```
